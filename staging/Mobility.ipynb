```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta
```

```python
mobility = pd.read_csv("Mobility.csv")
```

```python
mobility.head(100)
```

```python
s_date = datetime.strptime('20200901', '%Y%m%d')
e_date = datetime.strptime('20201231', '%Y%m%d')
mobility['date'] = pd.to_datetime(mobility['date'])
```

```python
mobility = mobility[(mobility['date'] >= s_date) & (mobility['date'] <= e_date)]
```

```python
mobility_filter = mobility[(mobility['sub_region_1'] == 'Ontario')]
```

```python
pd.set_option('display.max_rows', None)
```

```python
mobility_filter = mobility_filter[(mobility_filter['sub_region_2'].str.contains('Durham')) | (mobility_filter['sub_region_2'].str.contains('Halton')) | (mobility_filter['sub_region_2'].str.contains('Peel')) | (mobility_filter['sub_region_2'].str.contains('York')) | (mobility_filter['sub_region_2'].str.contains('Toronto')) | (mobility_filter['sub_region_2'].str.contains('Ottawa'))]
```

```python
mobility_filter = mobility_filter.filter(items=['sub_region_1', 'sub_region_2', 'metro_area', 'place_id',
                              'date', 'retail_and_recreation_percent_change_from_baseline', 
                              'grocery_and_pharmacy_percent_change_from_baseline', 
                              'parks_percent_change_from_baseline', 'transit_stations_percent_change_from_baseline',
                              'workplaces_percent_change_from_baseline', 'residential_percent_change_from_baseline'
                             ])
```

```python
mobility_filter
```

```python
Mobility_dimension = pd.DataFrame(columns=['mobility_id', 'Metro_area', 'Subregion_1', 'Subregion_2', 'Date',
                                     'Province', 'Retail_and_recreation', 'Grocery_and_pharmacy', 'Parks', 'Transit_stations', 'Workplaces',
                                          'Residential'])
```

```python
for idx, row in mobility_filter.iterrows():
    
    #Weather
    mobility_filter_row = [idx]
    mobility_filter_row.append(row["metro_area"])
    mobility_filter_row.append(row["sub_region_1"])
    if('Ottawa' in row["sub_region_2"]):
        mobility_filter_row.append('Ottawa')
    elif ('Durham' in row["sub_region_2"]):
        mobility_filter_row.append('Durham')
    elif ('Halton' in row["sub_region_2"]):
        mobility_filter_row.append('Halton')
    elif ('Peel' in row["sub_region_2"]): 
        mobility_filter_row.append('Peel')
    elif ('York' in row["sub_region_2"]):
        mobility_filter_row.append('York')
    elif ('Toronto' in row["sub_region_2"]):
        mobility_filter_row.append('Toronto')
    else :
        mobility_filter_row.append('unknown')
    mobility_filter_row.append(row["date"])
    mobility_filter_row.append('ON')
    mobility_filter_row.append(row["retail_and_recreation_percent_change_from_baseline"])
    mobility_filter_row.append(row["grocery_and_pharmacy_percent_change_from_baseline"])
    mobility_filter_row.append(row["parks_percent_change_from_baseline"])
    mobility_filter_row.append(row["transit_stations_percent_change_from_baseline"])
    mobility_filter_row.append(row["workplaces_percent_change_from_baseline"])
    mobility_filter_row.append(row["residential_percent_change_from_baseline"])
    Mobility_dimension.loc[len(Mobility_dimension)] = mobility_filter_row

Mobility_dimension.insert(0, "surrogate_key",  np.arange(len(Mobility_dimension)))
```

```python
Mobility_dimension
```

```python
Mobility_dimension["Date"] = pd.to_datetime(Mobility_dimension["Date"] ).dt.strftime('%Y-%m-%d')
```

```python
Mobility_dimension.to_csv("Mobility_dimension.csv",index=False)
```
