```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta
```

```python
weathers_Ottawa = pd.read_csv("2020_Ottawa.csv")
```

```python
weathers_Ottawa = weathers_Ottawa.loc[244:]
```

```python
weathers_Ottawa
```

```python
weather_dimension = pd.DataFrame(columns=['Weather_ID', 'City', 'Date', 'Daily_high_temperature', 'Daily_low_temperature', 'Rain_mm',
                                     'Snow_cm'])
```

```python
for idx, row in weathers_Ottawa.iterrows():
    
    #Weather
    weather_row = []
    weather_row.append(row["Climate ID"])
    weather_row.append('Ottawa')
    weather_row.append(row["Date/Time"])
    weather_row.append(row["Max Temp (°C)"])
    weather_row.append(row["Min Temp (°C)"])
    weather_row.append(row["Total Rain (mm)"])
    weather_row.append(row["Total Snow (cm)"])
    weather_dimension.loc[len(weather_dimension)] = weather_row
```

```python
weather_dimension
```

```python
weathers_Peel = pd.read_csv("2020_Peel.csv")
weathers_York = pd.read_csv("2020_York.csv")
weathers_Toronto = pd.read_csv("2020_Toronto.csv")
```

```python
weathers_Peel = weathers_Peel.loc[244:]
weathers_York = weathers_York.loc[244:]
weathers_Toronto = weathers_Toronto.loc[244:]
```

```python
for idx, row in weathers_Peel.iterrows():
    
    #Weather
    weather_row = []
    weather_row.append(row["Climate ID"])
    weather_row.append('Peel')
    weather_row.append(row["Date/Time"])
    weather_row.append(row["Max Temp (°C)"])
    weather_row.append(row["Min Temp (°C)"])
    weather_row.append(row["Total Rain (mm)"])
    weather_row.append(row["Total Snow (cm)"])
    weather_dimension.loc[len(weather_dimension)] = weather_row
    
for idx, row in weathers_York.iterrows():
    
    #Weather
    weather_row = []
    weather_row.append(row["Climate ID"])
    weather_row.append('York')
    weather_row.append(row["Date/Time"])
    weather_row.append(row["Max Temp (°C)"])
    weather_row.append(row["Min Temp (°C)"])
    weather_row.append(row["Total Rain (mm)"])
    weather_row.append(row["Total Snow (cm)"])
    weather_dimension.loc[len(weather_dimension)] = weather_row

for idx, row in weathers_Toronto.iterrows():
    
    #Weather
    weather_row = []
    weather_row.append(row["Climate ID"])
    weather_row.append('Toronto')
    weather_row.append(row["Date/Time"])
    weather_row.append(row["Max Temp (°C)"])
    weather_row.append(row["Min Temp (°C)"])
    weather_row.append(row["Total Rain (mm)"])
    weather_row.append(row["Total Snow (cm)"])
    weather_dimension.loc[len(weather_dimension)] = weather_row
```

```python
weather_dimension.insert(0, "surrogate_key",  np.arange(len(weather_dimension)))
```
