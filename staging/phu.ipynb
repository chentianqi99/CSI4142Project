```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta
```

```python
phu = pd.read_csv("patients.csv")
```

```python
s_date = datetime.strptime('20200901', '%Y%m%d')
e_date = datetime.strptime('20201231', '%Y%m%d')
phu['Case_Reported_Date'] = pd.to_datetime(phu['Case_Reported_Date'])
```

```python
phu = phu[(phu['Case_Reported_Date'] >= s_date) & (phu['Case_Reported_Date'] <= e_date)]
```

```python
phu_filter = phu[phu['Reporting_PHU'].str.contains("Peel") | phu['Reporting_PHU'].str.contains("York") | phu['Reporting_PHU'].str.contains("Durham") | phu['Reporting_PHU'].str.contains("Halton") | phu['Reporting_PHU_City'].str.contains("Toronto") | phu['Reporting_PHU_City'].str.contains("Ottawa")]
```

```python
phu_filter = phu_filter.filter(items=['Reporting_PHU_ID', 'Reporting_PHU', 'Reporting_PHU_Address', 'Reporting_PHU_City', 'Reporting_PHU_Postal_Code', 'Reporting_PHU_Website', 'Reporting_PHU_Latitude', 'Reporting_PHU_Longitude'])
```

```python
phu_filter.drop_duplicates(keep='first', inplace=True)
```

```python
phu_filter
```

```python
PHU_location_dimension = pd.DataFrame(columns=['PHU_location_key', 'PHU_name', 'Address',
                                     'City', 'Postal_code', 'Province', 'URL', 'Latitude', 'Longtitude'])
```

```python
for idx, row in phu_filter.iterrows():
    
    #Weather
    phu_filter_row = []
    phu_filter_row.append(row["Reporting_PHU_ID"])
    phu_filter_row.append(row["Reporting_PHU"])
    phu_filter_row.append(row["Reporting_PHU_Address"])
    phu_filter_row.append(row["Reporting_PHU_City"])
    phu_filter_row.append(row["Reporting_PHU_Postal_Code"])
    phu_filter_row.append('ON')
    phu_filter_row.append(row["Reporting_PHU_Website"])
    phu_filter_row.append(row["Reporting_PHU_Latitude"])
    phu_filter_row.append(row["Reporting_PHU_Longitude"])
    PHU_location_dimension.loc[len(PHU_location_dimension)] = phu_filter_row

PHU_location_dimension.insert(0, "surrogate_key",  np.arange(len(PHU_location_dimension)))
```
