```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta
```

```python
measure = pd.read_csv("special_measure.csv")
```

```python
measure = measure[measure['Reporting_PHU'].str.contains("Peel") | measure['Reporting_PHU'].str.contains("York") | measure['Reporting_PHU'].str.contains("Durham") | measure['Reporting_PHU'].str.contains("Halton") | measure['Reporting_PHU'].str.contains("Toronto") | measure['Reporting_PHU'].str.contains("Ottawa")]
```

```python
measure = measure.filter(items=['Reporting_PHU', 'Reporting_PHU_id', 'Status_PHU', 'start_date', 'end_date'])
```

```python
measure_dimension = pd.DataFrame(columns=['measure_id', 'Reporting_PHU', 'Reporting_PHU_id', 'Status_PHU', 'start_date', 'end_date'])
```

```python
for idx, row in measure.iterrows():
    
    #Weather
    measure_row = [idx]
    measure_row.append(row["Reporting_PHU"])
    measure_row.append(row["Reporting_PHU_id"])
    measure_row.append(row["Status_PHU"])
    measure_row.append(row["start_date"])
    measure_row.append(row["end_date"])
    measure_dimension.loc[len(measure_dimension)] = measure_row
```

```python
measure_dimension["start_date"] = pd.to_datetime(measure_dimension["start_date"] ).dt.strftime('%Y-%m-%d')
measure_dimension["end_date"] = pd.to_datetime(measure_dimension["end_date"] ).dt.strftime('%Y-%m-%d')
```

```python
measure_dimension.insert(0, "surrogate_key",  np.arange(len(measure_dimension)))
```
